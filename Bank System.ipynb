```python
class Account:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{self.name}, your new balance is: {self.balance}$")

    def withdraw(self, amount):
        if amount > self.balance:
            return False
        self.balance -= amount
        print(f"{self.name}, your new balance is: {self.balance}$")
        return True

    def check_balance(self):
        print(f"{self.name}, your balance is: {self.balance}$")


class Bank:
    def __init__(self, name):
        self.name = name
        self.accounts = {}
        self.tries = 3

    def welcome(self):
        print(f"Welcome to {self.name} bank")

    def create_account(self):
        name = input("Please enter your name: ")
        if name in self.accounts:
            print("Account already exists.")
            return
        try:
            balance = float(input("Enter initial balance: "))
        except ValueError:
            print("Invalid amount. Try again.")
            return
        self.accounts[name] = Account(name, balance)
        print(f"Account created for {name} with balance: {balance}$")

    def main_menu(self):
        while self.tries > 0:
            name = input("Enter your name: ")
            if name not in self.accounts:
                self.tries -= 1
                print("Invalid name. Try again.")
                if self.tries == 0:
                    print("Out of tries.")
                    return
            else:
                self.account_operations(self.accounts[name])
                return

    def account_operations(self, account):
        while True:
            option = input("Choose an option:\n(1)Check Balance\n(2)Deposit\n(3)Withdraw\n(4)Exit:\n")
            if option == "1":
                account.check_balance()
            elif option == "2":
                try:
                    amount = float(input("Enter amount to deposit: "))
                    account.deposit(amount)
                except ValueError:
                    print("Invalid amount.")
            elif option == "3":
                try:
                    amount = float(input("Enter amount to withdraw: "))
                    if not account.withdraw(amount):
                        self.tries -= 1
                        if self.tries > 0:
                            print(f"Insufficient funds. Tries left: {self.tries}")
                        else:
                            print("Out of tries.")
                            sys.exit()
                except ValueError:
                    print("Invalid amount.")
            elif option == "4":
                account.check_balance()
                print("We were glad to serve you. Goodbye!")
                sys.exit()
            else:
                self.tries -= 1
                if self.tries > 0:
                    print("Invalid option. Try again.")
                else:
                    print("Out of tries.")
                    sys.exit()

    def run(self):
        self.welcome()
        while self.tries > 0:
            option = input("Choose an option:\n(1)Account Creation\n(2)Main Menu:\n")
            if option == "1":
                self.create_account()
            elif option == "2":
                self.main_menu()
            else:
                self.tries -= 1
                print("Invalid option.")
                if self.tries == 0:
                    print("Out of tries.")


class BankSystem:
    def __init__(self):
        bank_names = ["Alahly", "Masr", "CIB", "HSBC"]
        probabilities = [0.25, 0.25, 0.25, 0.25]
        chosen_bank = random.choices(bank_names, weights=probabilities, k=1)[0]
        self.bank = Bank(chosen_bank)

    def start(self):
        self.bank.run()


if __name__ == "__main__":
    app = BankSystem()
    app.start()
```
